```python
# https://training.gismentors.cz/geopython-zacatecnik/pyproj/index.html
# https://pyproj4.github.io/pyproj/stable/gotchas.html#axis-order-changes-in-proj-6

from pyproj import CRS
from pyproj import Transformer
import pandas as pd
```

```python
# převodní transformace: S-JTSK (Křovák) -> WGS84
# pokud se ukáže, že hodnoty nejsou v EPSG:5514, stačí změnit první EPSG kód

def convert_coord(x:float,y:float):
    sjtsk = CRS("epsg:5514") # https://epsg.io/5514 S-JTSK / Krovak East North
    wgs = CRS("epsg:4326")  # https://epsg.io/4326 GPS

    transformer = Transformer.from_crs(sjtsk, wgs)
    lat, long = transformer.transform(-y, -x)
    return lat, long

# x = 1046949.0
# y = 540446.0
# lat, long = convert_coord(x,y)
# print(f'{lat=}, {long =}')
```

```python
# vstupní a výstupní soubor
INPUT = "STROMY_S-JTSK.csv"
OUTPUT = "STROMY_WGS84.csv"

# načtení dat
df = pd.read_csv(INPUT)
```

```python
# https://www.reddit.com/r/learnpython/comments/vlm67x/append_multiple_columns_applying_function_that/
# df[['c','d']] = df.apply(lambda x: pd.Series(f(x.a,x.b)), axis=1)
# df_trees[['lat', 'long']] = df_trees.apply(lambda x: get_lat_long(x.souradnice)[0], axis=1)

df[['lat', 'long']] = df.apply(lambda x: pd.Series(convert_coord(x["LONGITUDE_KROVAK"], x["LATITUDE_KROVAK"])), axis=1)
```

```python
df.to_csv(OUTPUT, index=False)
```
